```python
import pandas as pd
import numpy as np
import statistics as stats
```

```python
df1 = pd.read_csv(r'\applications(2.0).csv')
df2 = pd.read_csv(r'\industries(2.0).csv')
```

```python
df1.head(5)
```

```python
df2.head(5)
```

```python
df1 = df1.drop_duplicates(subset = 'applicant_id', keep = 'last')
df1['External Rating'] = df1['External Rating'].fillna(0)
df1['Education level'] = df1['Education level'].fillna('Середня')

df1.head(5)


```

```python
df_merg = df1.merge(df2, on = 'Industry', how = 'left')
df_merg
```

```python
df_merg['Rating'] = 0
df_merg['Rating'] += df_merg['Score']
df_merg.loc[df_merg['Age'].between(35, 55), 'Rating'] += 20
df_merg.loc[df_merg['Marital status'] == 'Married', 'Rating'] += 20
df_merg.loc[df_merg['Location'] == 'Київ чи область', 'Rating'] += 10
df_merg.loc[df_merg['External Rating'] >= 7.0, 'Rating'] += 20
df_merg.loc[df_merg['External Rating'] <= 2.0, 'Rating'] -= 20

df_merg['Applied at'] = pd.to_datetime(df_merg['Applied at'], dayfirst=True, errors='coerce')
df_merg.loc[df_merg['Applied at'].dt.weekday <= 4, 'Rating'] += 20
df_merg
```

```python
df_merg.shape
```

```python
df_filt = df_merg[df_merg['Rating'].between(0, 100)]
df_filt.shape
```

```python
df_filt.groupby(df_filt['Applied at'].dt.to_period('W'))['Rating'].mean().round(1)
```
